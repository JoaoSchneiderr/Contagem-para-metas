### 2. Gravidade dos acidentes por mês em Nova Iguaçu

Show! Agora que sabemos quantos acidentes por mês e quantos acidentes no total tiveram no ano de 2025, qual foi a gravidade deles?

```python
nova_iguacu_df['data_inversa'] = pd.to_datetime(nova_iguacu_df['data_inversa'])
nova_iguacu_df['mes_num'] = nova_iguacu_df['data_inversa'].dt.month
meses_pt = {
    1: 'Janeiro', 2: 'Fevereiro', 3: 'Março', 4: 'Abril',
    5: 'Maio', 6: 'Junho', 7: 'Julho', 8: 'Agosto',
    9: 'Setembro', 10: 'Outubro', 11: 'Novembro', 12: 'Dezembro'
}
nova_iguacu_df['mes'] = nova_iguacu_df['mes_num'].map(meses_pt)

gravidade_mensal = nova_iguacu_df.groupby('mes')[['mortos', 'feridos_graves', 'feridos_leves', 'ilesos']].sum()

gravidade_mensal = gravidade_mensal.reindex([
    'Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho',
    'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro'
], fill_value=0)

print("\nAnálise da Gravidade dos Acidentes por Mês em Nova Iguaçu:")
display(gravidade_mensal)

total_mortos_geral = gravidade_mensal['mortos'].sum()
total_feridos_graves_geral = gravidade_mensal['feridos_graves'].sum()
total_feridos_leves_geral = gravidade_mensal['feridos_leves'].sum()
total_ilesos_geral = gravidade_mensal['ilesos'].sum()

print("\n--- Totais Gerais (Ano de 2025) ---")
print(f"Total Geral de Mortos: {total_mortos_geral:.0f}")
print(f"Total Geral de Feridos Graves: {total_feridos_graves_geral:.0f}")
print(f"Total Geral de Feridos Leves: {total_feridos_leves_geral:.0f}")
print(f"Total Geral de Ilesos: {total_ilesos_geral:.0f}")
```

Com isso podemos afirmar que 48.42% dos acidentes em Nova Iguaçu tiveram vítimas (mortos, feridos leves ou graves).

51.58% dos acidentes não registraram vítimas.

Isso representa que a taxa de acidentes com mortos é extremamente baixa, sendo de apenas 2.16% e a média de vitimas por acidente(incluindo mortos e feridos) de apenas 0.48

### 3. Tipos de acidentes

Nesta seção, vamos explorar os tipos de acidentes que ocorrem com maior frequência em Nova Iguaçu e identificar aqueles que resultam em maior número de vítimas

```python
print("\n Tipos de Acidentes Mais Comuns")
tipos_acidentes_comuns = nova_iguacu_df['tipo_acidente'].value_counts().head(10)
display(tipos_acidentes_comuns)

gravidade_por_tipo_acidente = nova_iguacu_df.groupby('tipo_acidente')[['mortos', 'feridos_graves', 'feridos_leves']].sum()
gravidade_por_tipo_acidente['total_vitimas'] = gravidade_por_tipo_acidente['mortos'] + gravidade_por_tipo_acidente['feridos_graves'] + gravidade_por_tipo_acidente['feridos_leves']

tipos_acidentes_mais_graves = gravidade_por_tipo_acidente.sort_values(by='total_vitimas', ascending=False).head(10)

print("\n Tipos de Acidentes Mais Graves (Top 10 por Total de Vítimas)")
display(tipos_acidentes_mais_graves)

print("\n Classificação de Acidentes Mais Comuns")
classificacao_acidente_comum = nova_iguacu_df['classificacao_acidente'].value_counts()
display(classificacao_acidente_comum)
```

### 5. Perfil das pessoas envolvidas

Certo, sabemos muito mais agora! Porém, como eram as pessoas que estavam envolvidas nesses acidentes?

```python
df_perfil = nova_iguacu_df[['sexo', 'idade', 'estado_fisico', 'tipo_envolvido']].copy()
df_perfil['idade'] = pd.to_numeric(df_perfil['idade'], errors='coerce')
df_perfil = df_perfil.dropna(subset=['idade'])
display(df_perfil.head())
```

```python
bins = [0, 18, 25, 35, 45, 55, 65, 100]
labels = ['0-17', '18-24', '25-34', '35-44', '45-54', '55-64', '65+']

df_perfil['faixa_etaria'] = pd.cut(df_perfil['idade'], bins=bins, labels=labels, right=False)

print("\nDistribuição por Faixa Etária (Indivíduos Envolvidos em Acidentes):")
display(df_perfil['faixa_etaria'].value_counts().sort_index())
```

```python
df_vitimas = df_perfil[df_perfil['estado_fisico'].isin(['Ferido Grave', 'Ferido Leve', 'Óbito'])]
idade_media_vitimas = df_vitimas['idade'].mean()

print(f"Idade média das vítimas (Ferido Grave, Ferido Leve ou Morto): {idade_media_vitimas:.2f} anos")
```

```python
print("\nDistribuição por Sexo (Indivíduos Envolvidos em Acidentes):")
contagem_sexo = df_perfil['sexo'].value_counts()
display(contagem_sexo)

porcentagem_sexo = df_perfil['sexo'].value_counts(normalize=True) * 100
print("\nPorcentagem por Sexo (Indivíduos Envolvidos em Acidentes):")
display(porcentagem_sexo)
```

```python
print("\nContagem de Condutores vs. Passageiros:\n")
contagem_condutor_passageiro = df_perfil['tipo_envolvido'].value_counts().loc[['Condutor', 'Passageiro']]
display(contagem_condutor_passageiro)
```
